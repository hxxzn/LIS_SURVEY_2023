## 6) 공통함수 지정

**답변 정리 후 데이터 프레임 만들기**

```python
def count_to(series):

    # 시리즈 복사 - 원본 데이터 유지
    series_copy = series.copy()
    # 결측값 제거
    series_copy.dropna(inplace=True)
    # 시리즈를 리스트로 변환
    first_list = series_copy.tolist()

    # 쪼개야 할 단어들을 찾아서 리스트에서 제거하고 각 단어를 쪼갠 뒤 리스트에 추가
    words_to_split = [word for word in first_list if ',' in str(word)]
    for word in words_to_split:
        first_list.remove(word)
        first_list.extend(word.split(','))

    # 중복을 제거한 고유한 단어들의 리스트 생성
    unique_words = list(set(first_list))
    # 각 고유한 단어의 출현 빈도 리스트에 저장
    counts = [first_list.count(word) for word in unique_words]

    # 단어와 출현 빈도로 데이터프레임 생성
    count_df = pd.DataFrame({series.name: unique_words, '카운트': counts})
    # 출현 빈도에 따라 내림차순으로 데이터프레임 정렬
    count_df = count_df.sort_values('카운트', ascending=False).reset_index(drop=True)

    # 정렬된 데이터프레임 반환
    return count_df
```

**예외 처리 함수 - 같은 선택지 다르게 카운트 되는 경우**

```python
def count_to_outlier(df, column):
    # 공백을 제거한 리스트 생성
    tmp_list = [i.strip() for i in df[column]]
    # 데이터프레임의 열을 업데이트
    df[column] = tmp_list
    # 공백을 제거한 값에 기반하여 그룹화하고 합계 계산
    df = df.groupby(column).sum()

    return df
```

**구분 / 퍼센트 추가 함수**

```python
def add_classes(df, name, ex_column, new_column):
    # 데이터프레임의 인덱스 재설정
    df.reset_index(drop=True, inplace=True)
    # '구분' 열에 'name' 값을 할당
    df['구분'] = name
    per = []
    # 각 행에 대해 '카운트'의 비율을 계산하여 '퍼센트(%)' 열에 추가
    for i in range(0, len(df)):
        per.append(round(df['카운트'].loc[i] / df['카운트'].sum() * 100, 2))
    df['퍼센트(%)'] = per
    # 열 이름 변경
    df.rename(columns={ex_column: new_column}, inplace=True)
    # '퍼센트(%)' 열을 기준으로 내림차순 정렬
    df.sort_values(by='퍼센트(%)', ascending=False, inplace=True)
    return df
```

```python
def likert_graph(df, theme, order):

    # 퍼센트 값 계산
    a = pd.pivot_table( df, index='구분', columns=theme, values='퍼센트(%)')
    a = a[order]  # 순서 정렬
    a.sort_values(by='구분', ascending=False, inplace=True)

    # 카운트 계산
    b = pd.pivot_table( df, index='구분', columns=theme, values='카운트')
    b = b[order]
    b.sort_values(by='구분', ascending=False, inplace=True)

    # 그래프 색상 코드 지정
    colors = [px.colors.sequential.dense[2], px.colors.sequential.dense[3],
              px.colors.sequential.dense[4], px.colors.sequential.dense[5], px.colors.sequential.dense[6]]

    # flotly 생성
    fig = go.Figure()
    for column, color, bcol in zip(a.columns, colors, b.columns):
        fig.add_trace(go.Bar( x=a[column], y=a.index, name=column,
            customdata=b[[bcol]], orientation='h', marker_color=color,
            hovertemplate= "<b>{}:</b><br>".format(column) + "응답자수: %{customdata:,}명" +  "<extra></extra>"), )

    # 레이아웃 설정
    fig.update_layout(barmode='stack',
                      yaxis={'visible': True, 'showticklabels': True, 'title': ''},
                      xaxis={'visible': False, 'showticklabels': False}, height=350, showlegend=True,
                      plot_bgcolor='rgba(0, 0, 0, 0)', paper_bgcolor='rgba(0,0,0,0)', margin=dict(l=2, r=2),
                      legend=dict(yanchor="top", y=1.25, xanchor="left", x=0.0, orientation="h", font_size=12,
                                  traceorder='normal', itemsizing='trace'))
    fig.update_traces(texttemplate='%{x:.2f}%', textposition='inside', textfont_size=14)

    # 그래프 출력
    fig.show()

    # html 변환
    fig.write_html("%s.html" % theme)
```

```python
def likert_graph_order(df, theme, order):

    category_orders = {"구분": ["취업자", "취준생", "대학원재학생", "학부재학생"]}  # 순서 고정

    # 퍼센트 값 계산
    a = pd.pivot_table(df, index='구분', columns=theme, values='퍼센트(%)')
    a = a[order]  # 순서 정렬
    a = a.reindex(category_orders['구분'])  # 고정된 category_orders를 적용하여 원하는 순서로 재정렬
    a.sort_values(by='구분', ascending=False, inplace=True)

    # 카운트 계산
    b = pd.pivot_table(df, index='구분', columns=theme, values='카운트')
    b = b[order]
    b = b.reindex(category_orders['구분'])  # 고정된 category_orders를 적용하여 원하는 순서로 재정렬
    b.sort_values(by='구분', ascending=False, inplace=True)

    # 그래프 색상 코드 지정
    colors = [px.colors.sequential.dense[2], px.colors.sequential.dense[3],
              px.colors.sequential.dense[4], px.colors.sequential.dense[5], px.colors.sequential.dense[6]]

    # flotly 생성
    fig = go.Figure()
    for column, color, bcol in zip(order, colors, order):
        fig.add_trace(go.Bar(x=a[column], y=a.index, name=column,
                             customdata=b[[bcol]], orientation='h', marker_color=color,
                             hovertemplate="<b>{}:</b><br>".format(column) + "응답자수: %{customdata:,}명" + "<extra></extra>"), )

    # 레이아웃 설정
    fig.update_layout(barmode='stack',
                      yaxis={'categoryorder': 'array', 'categoryarray': category_orders['구분'],
                             'visible': True, 'showticklabels': True, 'title': ''},
                      xaxis={'visible': False, 'showticklabels': False}, height=350, showlegend=True,
                      plot_bgcolor='rgba(0, 0, 0, 0)', paper_bgcolor='rgba(0,0,0,0)', margin=dict(l=2, r=2),
                      legend=dict(yanchor="top", y=1.25, xanchor="left", x=0.0, orientation="h", font_size=12,
                                  traceorder='normal', itemsizing='trace'))
    fig.update_traces(texttemplate='%{x:.2f}%', textposition='inside', textfont_size=14)

    # 그래프 출력
    fig.show()

    # html 변환
    fig.write_html("%s.html" % theme)
```

# 4️⃣ **세부 분석**

## 1) 다전공에 따라

```python
# (직무에 따른) 세부 분석용 새로운 테이블 그리는 함수 -- 다전공이수종류별선택이유
def make_pv(name, index1, index2, type): #name에는 student, employed, prepare 중에서, index1은 기준컬럼, index2는 y축컬럼, type은 재학생, 취업자, 취준생 중에서
  M=pd.pivot_table(name, index=[index1, index2], values=['index'], aggfunc='count')
  M['구분']=type
  M.reset_index(inplace=True)
  M.rename(columns={'index':'카운트'}, inplace=True)
  return M
```

```python
# (직무에 따른) 세부 분석용 새로운 테이블 그리는 함수 -- 다전공별선택이유
def make_pv2(name, index1, index2, index3, type): #name에는 student, employed, prepare 중에서, index1은 기준컬럼, index2는 y축컬럼, type은 재학생, 취업자, 취준생 중에서
  M=pd.pivot_table(name, index=[index1, index2, index3], values=['index'], aggfunc='count')
  M['구분']=type
  M.reset_index(inplace=True)
  M.rename(columns={'index':'카운트'}, inplace=True)
  return M
```

```python
# 다전공이수종류별 이유 함수
def multi_class_reason(df, col1, col2, c) :
  ndf = make_pv(df, col1, col2, c)
  ndf.rename(columns={col1:'다전공이수종류', col2:'다전공선택이유'}, inplace=True)

  del_list = []
  for i in range(len(ndf)):
    if ',' in ndf.iloc[i,1]:
      tmp = []
      del_list.append(i)
      tmp = ndf.iloc[i,1].split(',')
      for j in range(len(tmp)):
        app = []
        app = [ndf.iloc[i,0],tmp[j],1,c]
        ndf.loc[len(ndf)] = app
      #ndf.reset_index(inplace=True, drop = True)
    else: continue

  ndf = ndf.drop(del_list, axis=0, errors = 'raise')

  ndf = ndf.groupby(["다전공이수종류", "다전공선택이유"])['카운트'].sum()
  ndf = pd.DataFrame(ndf)
  ndf = ndf.reset_index()

  return ndf
```

```python
# 다전공별 이유 함수
def multi_reason(df, col1, col2, col3, col3_name, c) :
  ndf = make_pv2(df, col1, col2, col3, c)
  ndf.rename(columns={col1:'다전공이수종류', col3:col3_name}, inplace=True)

  del_list = []
  for i in range(len(ndf)):
    if ',' in ndf.iloc[i,2]:
      tmp = []
      del_list.append(i)
      tmp = ndf.iloc[i,2].split(',')
      for j in range(len(tmp)):
        app = []
        app = [ndf.iloc[i,0], ndf.iloc[i,1], tmp[j].strip(),1,c]
        ndf.loc[len(ndf)] = app
      #ndf.reset_index(inplace=True, drop = True)
    else: continue

  ndf = ndf.drop(del_list, axis=0, errors = 'raise')

  ndf = ndf.groupby(["다전공이수종류", "다전공", col3_name])['카운트'].sum()
  ndf = pd.DataFrame(ndf)
  ndf = ndf.reset_index()

  return ndf
```

```python
# 데이터프레임 카운트 횟수 정리해주는 함수
def multi_sum(df, col1, col2, col3) :
  ndf = df.loc[:, col1:col3]
  ndf = pd.DataFrame(ndf.groupby([col1, col2])[col3].sum())
  ndf.reset_index(inplace=True)

  return ndf
```

```python
#세부 분석용 새로운 테이블 그리는 함수
def year_df(name,index1, index2): #name은 student, prepare, employed index 은 보고싶은 컬럼값
  year_df=pd.pivot_table(name, index=[index1, index2], values=['index'], aggfunc='count')
  year_df.rename(columns={'index':'카운트'}, inplace=True)
  year_df.reset_index(inplace=True)
  year_df=pd.pivot_table(year_df, index=[index1, index2], values=[ '카운트'])
  year_df.reset_index(inplace=True)
  return year_df
```

```python
  def count_to_outlier2(df, column, index1):
    # 공백을 제거한 리스트 생성
    tmp_list = [i.strip() for i in df[column]]
    # 데이터프레임의 열을 업데이트
    df[column] = tmp_list
    # 공백을 제거한 값에 기반하여 피벗테이블 형성
    df = pd.pivot_table(df, index=[column, index1], values=['카운트'])
    df.reset_index(inplace=True)
    return df
```
