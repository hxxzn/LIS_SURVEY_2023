## 공통함수 지정

**답변 정리 후 데이터프레임 만드는 함수**

```python
def count_to(series, new_col):
    series_copy = series.copy()
    # 결측값 제거
    series_copy.dropna(inplace=True)
    # 시리즈를 리스트로 변환
    first_list = series_copy.tolist()

    # 쪼개야 할 단어들을 찾아서 리스트에서 제거하고 각 단어를 쪼갠 뒤 리스트에 추가
    words_to_split = [word for word in first_list if ',' in str(word)]
    for word in words_to_split:
        first_list.remove(word)
        first_list.extend(word.split(','))

    # 중복을 제거한 고유한 단어들의 리스트 생성
    unique_words = list(set(first_list))
    # 각 고유한 단어의 출현 빈도를 계산하여 리스트에 저장
    counts = [first_list.count(word) for word in unique_words]

    # 고유한 단어와 해당 단어의 출현 빈도로 데이터프레임 생성
    count_df = pd.DataFrame({new_col: unique_words, '카운트': counts})
    # 출현 빈도에 따라 내림차순으로 데이터프레임 정렬
    count_df = count_df.sort_values('카운트', ascending=False).reset_index(drop=True)
    # 정렬된 데이터프레임 반환
    return count_df
```

**예외 처리 함수 - 같은 선택지 다르게 카운트 되는 경우**

```python
#같은 선택지인데 다르게 카운트 되는 부분이 있을 경우 예외 처리하는 함수
def count_to_outlier(df, column):
  tmp_list=[]
  for i in df[column]:
    tmp_list.append(i.strip())
  df[column] = tmp_list
  df = df.groupby(column).sum()
  return df
```

**구분 / 퍼센트 추가 함수**

```python
def add_classes(df, name, ex_column, new_column):
    # 데이터프레임의 인덱스 재설정
    df.reset_index(drop=True, inplace=True)
    # '구분' 열에 'name' 값을 할당
    df['구분'] = name
    per = []
    # 각 행에 대해 '카운트'의 비율을 계산하여 '퍼센트(%)' 열에 추가
    for i in range(0, len(df)):
        per.append(round(df['카운트'].loc[i] / df['카운트'].sum() * 100, 2))
    df['퍼센트(%)'] = per
    # 열 이름 변경
    df.rename(columns={ex_column:new_column}, inplace=True)
    # '퍼센트(%)' 열을 기준으로 내림차순 정렬
    df.sort_values(by='퍼센트(%)', ascending=False, inplace=True)
    return df
```

**데이터프레임 기반 차트 출력 함수 - 레이아웃 미지정**

```python
def create_chart(dfs, theme):

    # 그래프 스타일 설정
    new = [px.colors.sequential.deep[0], px.colors.sequential.deep[1], px.colors.sequential.deep[2],
           px.colors.sequential.deep[3], px.colors.sequential.deep[4], px.colors.sequential.deep[5],
           px.colors.sequential.deep[8], px.colors.sequential.deep[9]]

    # 그래프 생성
    fig = px.bar(dfs, x="퍼센트(%)", y="구분", color=theme, orientation='h', text='퍼센트(%)',
                 height=300, color_discrete_sequence=new, custom_data=['카운트', theme])

    # 레이아웃 설정
    fig.update_layout(xaxis_tickfont_size=10,
                      legend=dict(yanchor="top", y=1.35, xanchor="left", x=0.0, orientation="h", font_size=10),
                      yaxis={'visible': True, 'showticklabels': True, 'title': ''},
                      xaxis={'visible': False, 'showticklabels': False},
                      showlegend=True, font=dict(size=10, color="black"), margin=dict(l=2, r=2),
                      plot_bgcolor='rgba(0, 0, 0, 0)', paper_bgcolor='rgba(0,0,0,0)')

    # 표시되는 값 설정
    fig.update_traces(texttemplate='%{text:.2f}%', textposition='inside', textfont_size=14)

    # 호버링 설정
    fig.update_traces(hovertemplate='<br>%{customdata[1]}</br>응답 건수: %{customdata[0]} 건 <br>')

    # 그래프 출력
    fig.show()

    # html 변환
    fig.write_html("%s.html" % theme)
```
