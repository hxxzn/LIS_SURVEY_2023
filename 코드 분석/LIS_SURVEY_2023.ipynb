## 함수 설정

답변 정리 후 데이터 프레임 만들기

```python
def count_to(series, name):

    # 시리즈 복사 - 원본 데이터 유지
    series_copy = series.copy()
    # 결측값 제거
    series_copy.dropna(inplace=True)
    # 시리즈를 리스트로 변환
    first_list = series_copy.tolist()

    # 쪼개야 할 단어들을 찾아서 리스트에서 제거하고 각 단어를 쪼갠 뒤 리스트에 추가
    words_to_split = [word for word in first_list if ',' in str(word)]
    for word in words_to_split:
        first_list.remove(word)
        first_list.extend(word.split(','))

    # 중복을 제거한 고유한 단어들의 리스트 생성
    unique_words = list(set(first_list))
    # 각 고유한 단어의 출현 빈도 리스트에 저장
    counts = [first_list.count(word) for word in unique_words]

    # 단어와 출현 빈도로 데이터프레임 생성
    count_df = pd.DataFrame({name: unique_words, '카운트': counts})
    # 출현 빈도에 따라 내림차순으로 데이터프레임 정렬
    count_df = count_df.sort_values('카운트', ascending=False).reset_index(drop=True)

    # 정렬된 데이터프레임 반환
    return count_df
```

예외 처리 함수 - 같은 선택지 다르게 카운트 되는 경우

```python
def count_to_outlier(df, column):
    # 공백을 제거한 리스트 생성
    tmp_list = [i.strip() for i in df[column]]
    # 데이터프레임의 열을 업데이트
    df[column] = tmp_list
    # 공백을 제거한 값에 기반하여 그룹화하고 합계 계산
    df = df.groupby(column).size().reset_index(name='카운트')

    return df
```

구분 / 퍼센트 추가 함수

```python
def add_classes(df, name, ex_column, new_column):
    # 데이터프레임의 인덱스 재설정
    df.reset_index(drop=True, inplace=True)
    # '구분' 열에 'name' 값을 할당
    df['구분'] = name
    per = []
    # 각 행에 대해 '카운트'의 비율을 계산하여 '퍼센트(%)' 열에 추가
    for i in range(0, len(df)):
        per.append(round(df['카운트'].loc[i] / df['카운트'].sum() * 100, 2))
    df['퍼센트(%)'] = per
    # 열 이름 변경
    df.rename(columns={ex_column: new_column}, inplace=True)
    # '퍼센트(%)' 열을 기준으로 내림차순 정렬
    df.sort_values(by='퍼센트(%)', ascending=False, inplace=True)
    return df
```

**매우 그렇다 - 매우 아니다 유형 그래프 함수**

```python
import plotly.graph_objects as go
```

```python
def likert_graph(df, theme, order):

    # 퍼센트 값 계산
    a = pd.pivot_table(
        df,
        index='구분',
        columns=theme,
        values='퍼센트(%)')
    a = a[order]  # 순서 정렬
    a.sort_values(by='구분', ascending=False, inplace=True)

    # 카운트 계산
    b = pd.pivot_table(
        df,
        index='구분',
        columns=theme,
        values='카운트')
    b = b[order]
    b.sort_values(by='구분', ascending=False, inplace=True)

    # 그래프 색상 코드 지정
    colors = ['#FFCE9A', '#FFC4B6', '#FAC1C8', '#D9C0CB', '#BBABC3', '#929AB8', '#628BA6', '#337C89']

    # flotly 생성
    fig = go.Figure()
    for column, color, bcol in zip(a.columns, colors, b.columns):
        fig.add_trace(go.Bar(
            x=a[column],
            y=a.index,
            name=column,
            customdata=b[[bcol]],
            orientation='h',
            marker_color=color,
            hovertemplate=
            "<b>{}:</b><br>".format(column) +
            "응답자수: %{customdata:,}명" +
            "<extra></extra>"), )

    # 레이아웃 설정
    fig.update_layout(barmode='stack',
                      yaxis={'visible': True, 'showticklabels': True, 'title': ''},
                      xaxis={'visible': False, 'showticklabels': False}, height=350,
                      showlegend=True,
                      plot_bgcolor='rgba(0, 0, 0, 0)', paper_bgcolor='rgba(0,0,0,0)', margin=dict(l=2, r=2),
                      legend=dict(yanchor="top", y=1.25, xanchor="left", x=0.0, orientation="h", font_size=12))
    fig.update_traces(texttemplate='%{x:.2f}%', textposition='inside', textfont_size=14)

    # 그래프 출력
    fig.show()

    # html 변환
    fig.write_html("%s.html" % theme)
```

**Scatter 그래프 함수**

```python
def scatter_plot(df, x_axis, y_axis, size_num, name):

    # 그래프 생성
    fig = px.scatter(df, x=x_axis, y=y_axis, color=x_axis, size='카운트', size_max=size_num,
                     color_discrete_sequence=px.colors.sequential.deep[2:], custom_data=['카운트'])

    # 레이아웃 설정
    fig.update_layout(height=500, width=700,
                      xaxis=dict(title='', showgrid=True, zeroline=True, showline=True, showticklabels=True),
                      yaxis=dict(title='', showgrid=True, zeroline=True, showline=True, showticklabels=True),
                      autosize=False, showlegend=False, plot_bgcolor='rgba(0,0,0,0)',
                      margin=dict(l=2, r=4), paper_bgcolor='rgba(0,0,0,0)',
                      legend=dict(yanchor="top", y=1.05, xanchor="left", x=0.0, orientation="h", font_size=12))

    # 축 설정
    fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='rgba(0,0,0,0)')
    fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='rgba(0,0,0,0)')

    # 호버링
    fig.update_traces(hovertemplate='<br>%{x|%Y/%m}</br>%{y}</br>응답 건수: %{customdata[0]} 건 <br>')

    # 출력
    fig.show()

    # html 변환
    fig.write_html("%s.html" % name)
```
